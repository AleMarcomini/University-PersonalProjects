## Exercise 1 - Blood disease screening method 
A well established and diffused method for detecting a disease in blood fails to detect the presence of disease in $15\%$ of the patients that actually have the disease.

A young UniPD startUp has developed an innovative method of screening. During the qualification phase, a random sample of $n = 75$ patients known to have the disease is screened using the new method.

#### A. What is the probability distribution of $y$, the number of times the new method fails to detect the disease?
The probability distribution of $y$ is a Binomial distribution. In fact, given a patient who has the disease only two mutually exclusive outputs are possible: either the test manages to detect the disease or it fails. In particular, it is possible to encode the "success" outcome as the realization of the event: "the test fails to detect the desease". 

#### B. On the $n = 75$ patients sample, the new method fails to detect the disease in $y = 6$ cases. What is the frequentist estimator of the failure probability of the new method?

In the frequentist approach an unbiased estimator of the failure probability is given, in the case of a single experiment run, by the number of successes divided by the total number of trials. The variance associated to the estimation is $p(1-p)/n$ (which should be estimated in the true value). Hence, for the case under investigation:

$$
p_{fail} = \frac{\# \ failures}{\# \ trials} = \frac{6}{75} = 8.0\%
$$

With a standard deviation of $\sigma = 3.1\%$.

#### C. Setup a bayesian computation of the posterior probability, assuming a Beta distribution with mean value $0.15$ and standard deviation $0.14$. Plot the posterior distribution for $y$, and mark on the plot the mean value and variance.

First of all, to perform calculation is necessary to turn the given mean and standard deviation into $\alpha$ and $\beta$ parameters of the Beta distribution. Thus, recalling from the theory that:

$$
\left\{ \begin{array}{2} 
\mu = \frac{\alpha}{\alpha+\beta} \\
\sigma^2 = \frac{\alpha\beta}{{(\alpha+\beta)}^2(\alpha+\beta+1)} 
\end{array} \right.
\iff
\left\{ \begin{array}{2} 
\alpha = \left(\frac{1 - \mu}{\sigma^2} - \frac{1}{\mu}\right)\mu^2\\
\beta = \alpha \left(\frac{1}{\mu}-1\right)
\end{array} \right.
$$

Hence, substituting $\mu = 0.15$ and $\sigma = 0.14$ it is straightforward to obtain $\alpha \simeq 0.83$ and $\beta \simeq 4.68$. 

Given the fact that the Beta distribution is the conjugate prior of a Binomial distribution, from the theory it is known that the posterior is a Beta distribution with parameters $\alpha' = \alpha + y = 6.83$, $\beta' = \beta + n - y = 73.68$, where $n$ and $y$ are the number of tests and wrong detection, respectively.

Finally, the mean and the variance for the posterior are easily computable from the equations above.

```r
#Beta prior statistical indexes
mu.prior = 0.15
sigma.prior = 0.14

#Beta prior paramters
alpha.prior = (((1-mu.prior)/(sigma.prior**2))-(1/mu.prior))*mu.prior**2
beta.prior = alpha.prior*((1/mu.prior)-1)

#Experiment outcomes
n = 75
y = 6

#Beta posterior parameters
alpha.post = alpha.prior + y
beta.post = beta.prior + n - y

#Beta posterior statistical indexes
mu.post = alpha.post/(alpha.post+beta.post)
sigma.post = sqrt((alpha.post*beta.post)/((alpha.post+beta.post+1)*(alpha.post+beta.post)**2))

#Frequentist estimator
p.freq = 6/75

#Plot
options(repr.plot.width=12, repr.plot.height=6)

p = seq(0,0.3,len=201)
one.sigma.p = seq(mu.post-sigma.post,mu.post+sigma.post,len=137)
prior = dbeta(p,alpha.prior,beta.prior)
post = dbeta(p,alpha.post,beta.post)

ylim=c(0,15)
xlim=c(0,0.3)
plot(p,prior,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,ylab='PDF',
     xlab='p',main='Prior, posterior and esimators', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 2, lwd = 3)
par(new=TRUE)
plot(p,post,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,col='red',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
abline(v=mu.post,lty=3,lwd=2,col='red')
abline(v=p.freq,lty=3,lwd=2,col='black')
polygon(c(one.sigma.p,mu.post+sigma.post, mu.post-sigma.post), 
        c(dbeta(one.sigma.p,alpha.post,beta.post), 0, 0),
        col=adjustcolor("red",alpha.f=0.2))
legend('topright', c('Prior','Posterior','Frequentist estimator'), col=c("navy", "red", "black"), lty=c(2,1,3))
```

```r
sprintf('The estimator of the frequentist approach is %.2f%%, the estimator in the Bayesian case is %.2f%% with a standard deviation of %.2f%%.', 
        100*p.freq,100*mu.post, sigma.post*100)
```

The plot above reports the prior and the posterior for the case under review. The red area corresponds to the $[\mu-\sigma,\mu+\sigma]$ interval of the distribution. As it looks clear, the frequentist approach estimator differs noticeably from the Bayesian one. Anyway, the two values remain compatible within one sigma.

#### D. Perform a test of hypothesis assuming that if the probability of failing to the detect the desease in ill patients is greater or equal than $p^*=15\%$, the new test is no better that the traditional method. Test the sample at a $\alpha=5\%$ level of significance in the Bayesian way.

Firstly, let me define the null hypothesis as follows:

$$
H_0 : p \ge p^* =  15\%
$$

This corresponds to the statement: "the new method is no better than the traditional one". Of course, this leads to the alternative hypothesis "the new method is better":

$$
H_1 = \bar{H}_0 : p < p^* = 15\%
$$

To perform an HT in a Bayesian way, I need to compute the posterior probability of the null hypothesis as the integral of the posterior $g$:

$$
P\left(H_0 : p \ge p^*\right|y) = \int_{p^*}^1g(p|y) \ dp = 1 - CDF(p^*)
$$

Hence, I reject the null hypothesis at a level of significance $\alpha$ if the probability associated to the null hypothesis is smaller than $\alpha$.

```r
#Define statistical params for plot
p.star = 0.15
a = 0.05
p.alpha = qbeta(1-a,alpha.post,beta.post)
P.H0 = 1 - pbeta(p.star,alpha.post,beta.post)

#Curves and polygons
x.pol1 = seq(0,p.alpha,len=151)
x.pol2 = seq(p.alpha,0.3,len=101)
post = dbeta(p,alpha.post,beta.post)
c.post = pbeta(p,alpha.post,beta.post)

#Plot
ylim=c(0,1)
xlim=c(0,0.3)
plot(p,1-c.post,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,ylab='1 - CDF',
     xlab='p',main='Bayesian hypothesis test', cex.lab=1.5, cex.main=2,
     col='black',type='l', lty = 1, lwd = 3)
abline(h=a,lwd=3,lty=1,col='blue')
arrows(p.star, 0.6, p.star, 1-pbeta(p.star,alpha.post,beta.post), col='red',lwd=3)
polygon(c(x.pol2,0.3,0,0,p.alpha), 
        c(1-pbeta(x.pol2,alpha.post,beta.post), 0, 0,a,a),
        col=adjustcolor("red",alpha.f=0.2))
polygon(c(x.pol1,p.alpha,0), 
        c(1-pbeta(x.pol1,alpha.post,beta.post), a, a),
        col=adjustcolor("green",alpha.f=0.2))
legend('topright', c('Acceptance','Rejection'), fill=c("green", "red"), title='Region')
```

The plot above reports the $1-CDF$ curve. The blue line corresponds to the $5\%$ significance level and divides the curve in an acceptance region and a rejection region. As we can see, the red arrow indicates the position of $p^*$ on the horizontal axis: the corresponding value of $1-CDF$ lies in the rejection region. As I result, in the Bayesian approach I can reject the null hypothesis at a $5\%$ level of significance.

The plot below tells the same thing in another way: since I reject the null hypothesis is $P(H_0) < \alpha$, hence I can write:

$$
P(H_0) < \alpha \implies \int_{p^*}^1g(p|y) \ dp <  \int_{p_{\alpha}}^1g(p|y) \implies p^* > p_{\alpha}
$$

This is true since $g(p|y)$ is positive definite. Above I named $p_{\alpha}$ the value of $p$ such that $1-CDF(p) = \alpha$. Hence, I make a plot directly on the PDF while indicating as threshold between the acceptance and rejection regions the value $p_{\alpha}$ and looking if $p^*$ lies below or above.

```r
#Polygons
x.pol1 = seq(0,p.alpha,len=151)
x.pol2 = seq(p.alpha,0.3,len=101)

#Plot
ylim=c(0,15)
xlim=c(0,0.3)
plot(p,post,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,ylab='PDF',
     xlab='p',main='Bayesian hypothesis test', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 1, lwd = 3)
arrows(p.star, 10, p.star, dbeta(p.star,alpha.post,beta.post), col='red',lwd=3)
polygon(c(x.pol2,0.3,p.alpha), 
        c(dbeta(x.pol2,alpha.post,beta.post), 0, 0),
        col=adjustcolor("red",alpha.f=0.2))
polygon(c(x.pol1,p.alpha,0), 
        c(dbeta(x.pol1,alpha.post,beta.post), 0, 0),
        col=adjustcolor("green",alpha.f=0.2))
legend('topright', c('Acceptance','Rejection'), fill=c("green", "red"), title='Region')
```

As expected, the null hypothesis can be rejected at this level of significance. In the plot above the value of $p_{\alpha}$ is $\simeq 14.1\%$ (smaller than $p^*$).

#### E. Perform the same hypothesis test in the classical frequentist way.

The null hypothesis and its negation are the same as above.
The null distribution of the test statistic is equal to the sampling distribution of the test statistic, given that the null hypothesis is true:

$$
P(y|H_0) = 
Binom(y|n,p=15\%)
$$

While setting the level of significance $\alpha=5\%$, using the Neyman and Pearson's approach I define as the rejection region for the null hypothesis the one for which the CDF is up to $\alpha$. While performing this operation, one must recall that the binomial distribution is a discrete PDF. As a consequence, there may be not a value of $y$ such as $CDF(y) = \alpha$. Hence, I find the closest value $\tilde\alpha$ to $\alpha$ such that it can be written as the CDF of a certain value. For the case under review, the best choice is $\tilde\alpha = 5.4\%$ corresponding to the probability of $y \le 6$. The interval $[0,6]$ is then chosen as the rejection region.

Graphically:

```r
#Data for statistical analysis
n = 75
p.star = 0.15
a = 0.05
k = 6

#Define axes for plot and threshold for CDF=alpha
x = 0:30
acc.region = 7:30
y = pbinom(x,size=n,prob=p.star)
th = qbinom(a,size=n,prob=p.star)

#Plot
plot(x-0.5,y,type='h',xaxs='i',yaxs='i',lwd=0, xlab='y', ylab='CDF', main='Null hypothesis testing', cex.main=1.5)
for (i in 0:29){
    if (i == k){rect(i-0.5, 0, i+0.5, y[i+1], col='red')}
    else if (i %in% acc.region){rect(i-0.5, 0, i+0.5, y[i+1], col=adjustcolor("green",alpha.f=0.3))}
    else rect(i-0.5, 0, i+0.5, y[i+1], col=adjustcolor("red",alpha.f=0.3))
}
abline(h=pbinom(k,size=n,prob=p.star), lty=2, lwd=2, col='blue')
```

As the plot above depicts, the number of observations $y=6$ falls at the very end of the rejection region (highlighted red color), yet inside it. Hence, also in the frequentist approach I reject the null hypothesis at the given level of significance.

## Exercise 2 - Horse-related deaths in the Prussian army

Ladislaus Josephovich Bortkiewicz was a Russian economist and statistician. He noted that the Poisson distribution can be very useful in applied statistics when describing low-frequency events in a large population. In a famous example he showed that the number of deaths by horse kick among the Prussian army follows the Poisson distribution.

Consider the following to sets of observations taken over a fixed large time interval in two different corps:

| $y$ death soldiers | 0 | 1 | 2 | 3 | 4 | > 4 |

| $n_1$ observations | 109 | 65 | 22 | 3 | 1 | 0 |

#### A. Assuming a uniform prior, compute and plot the posterior distribution for λ, the death rate over the measurement time. Determine the posterior mean, median and variance, and compute the 95% credibility interval.

From the theory, it is known that a Poissonian process with uniform (improper) prior $g(\mu) = 1$ for $\mu>0$ leads to a Gamma posterior $\Gamma(\alpha, \lambda)$ after $n$ observations $\{y_k\}_{k=1..n}$ with parameters $\alpha = \sum_k y_k + 1$ and $\lambda = n$. Thus for the first case: 

$$\left\{\begin{array}{2}
\alpha_1 = (109 \cdot 0 + 65 \cdot 1 + 22 \cdot 2 + 3 \cdot 3 + 1 \cdot 4) + 1 = 123  \\
\lambda_1 = 109+65+22+3+1 = 200
\end{array}\right.
\implies P(\mu|\{y_k\}_k,M) = \Gamma(123,200)
$$

While for the second set of observations:

$$\left\{\begin{array}{2}
\alpha_1 = (144 \cdot 0 + 91 \cdot 1 + 32 \cdot 2 + 11 \cdot 3 + 2 \cdot 4) + 1 = 197  \\
\lambda_1 = 144+91+32+11+8 = 280
\end{array}\right.
\implies P(\mu|\{y_k\}_k,M) = \Gamma(197,280)
$$

```r
#Define parameters
xmin = 0
xmax = 1.5
len = 501
x = seq(xmin,xmax,len=len)
alpha.1 = 123
lambda.1 = 200
alpha.2 = 197
lambda.2 = 280

#Define pdfs
prior <- rep(1,len)
post.1 <- dgamma(x,shape=alpha.1, rate=lambda.1)
post.2 <- dgamma(x,shape=alpha.2, rate=lambda.2)

#Plot
ylim=c(0,8.5)
options(repr.plot.width=10, repr.plot.height=6)
plot(x,prior,xaxs='i',yaxs='i',ylim=ylim,ylab='PDF',
     xlab='µ',main='Prior and posteriors', cex.lab=1.5, cex.main=2,
     col='black',type='l', lty = 2, lwd = 3)
par(new=TRUE)
plot(x,post.1,xaxs='i',yaxs='i',ylim=ylim,col='red',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(x,post.2,xaxs='i',yaxs='i',ylim=ylim,col='blue',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
legend('right', c('Prior','Posterior 1', "Posterior 2"), col=c("black", "red", 'blue'), lty=c(2,1,1))
```

From the plot above we can see how the two posteriors are really narrow, yet also noticeably different from each other.

#### A. Indicating with $p$ the probability that a sample of water taken from the stream has a high bacter $X$ level, find the frequentist estimator for $p$.

The process is clearly binomial (either the sample has high bacter level, or it has not). The "success" here is the presence of a high bacter level. #### D. Test the hypotesis:
$$
H_0 : p=p^* = 0.1 \mbox{ versus } H_1 :p\neq p^*
$$
#### at $5\%$ level of significance with both the frequentist and bayesian approach.
This is a two-tailed hypothesis test.

In the Bayesian framework all the information about the $p$ parameter are kept in the posterior. Since the probability related to the null hypothesis is clearly $0$ (punctual probability), to test $H_0$ at significance level $\alpha$ I compute a $(1-\alpha) \times 100\%$ credible interval for $p$, determined by the upper and lower $\alpha/2$ quantiles, and look for the position of $p^* = 0.1$: if it lies inside the credible interval, then the null hypothesis can not be rejected at this level of significance.

In particular, for $\alpha=5\%$ the credibile interval concides with the one calculated above. 

```r
p.star = 0.1

#Credibility intervals and posteriors for plot
CI = seq(CI.min,CI.max,len=201)
CI.post = dbeta(CI,alpha.post,beta.post)

p = seq(0,0.3,len=201)
post = dbeta(p,alpha.post,beta.post)

ylim=c(0,16)
xlim=c(0,0.3)
plot(p,post,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,ylab='PDF',
     xlab='p',main='Bayesian hypothesis test', cex.lab=1.5, cex.main=2,
     type='l', lty = 1, lwd = 3)
polygon(c(CI,CI.max,CI.min), 
        c(CI.post, 0, 0),
        col=adjustcolor("green",alpha.f=0.25))
polygon(c(seq(0,CI.min,len=101),CI.min,0), 
        c(dbeta(seq(0,CI.min,len=101),alpha.post,beta.post), 0, 0),
        col=adjustcolor("red",alpha.f=0.25))
polygon(c(seq(CI.max,1,len=101),1,CI.max), 
        c(dbeta(seq(CI.max,1,len=101),alpha.post,beta.post), 0, 0),
        col=adjustcolor("red",alpha.f=0.25))
text(x=0.09, y=13, 'C.I.')
arrows(p.star, 10, p.star, 0, col='red',lwd=3)
```

Hence, since $p^*$ falls in the acceptance region we can not reject the null hypothesis at level of significance $\alpha$.

As for the frequentist approach, the null distribution is the sampling distribution of y given $H_0$, i.e.: $P(y|n,p) = Binom(y|n = 116, p= p^*)$. While defining the rejection region, one must recall that the binomial is a discrete ditribution, hence it will be necessary to find a value $\tilde{\alpha}$ which is as close as possible to $\alpha$ without being impossible to reach by summing probabilities of $y \sim P(y|n,p)$.

After some trials, I decided to set $\tilde{\alpha} \simeq 4.74\%$, which is the CDF of $\left\{y \le 5\right\} \cup \left\{y \ge 19\right\}$. The latter becomes thus the rejection region.

```r
#Model parameters
n = 116
p.star = 0.1
k =11

#Define axes for plot and threshold for CDF=alpha
x = 0:25
y = dbinom(x,size=n,prob=p.star)
acc.region = 6:18

#Plot
plot(x-0.5,y,type='h',xaxs='i',yaxs='i',lwd=0, xlab='y', ylab='PDF', main='Null hypothesis testing', cex.main=1.5)
for (i in 0:25){
    if (i %in% acc.region){
        if (i == k){rect(i-0.5, 0, i+0.5, y[i+1], col="green")}
        else rect(i-0.5, 0, i+0.5, y[i+1], col=adjustcolor("green",alpha.f=0.3))
    }
    else rect(i-0.5, 0, i+0.5, y[i+1], col=adjustcolor("red",alpha.f=0.3))
}

sprintf('Rejection area: %.3f',sum(y[0:5]) + sum(y[19:25]))
```

Also from the frequentist point of view, the observations found fall into the acceptance region, hence we can not reject the null hypothesis.

### A new measurement, performed one month later on $n = 165$ water samples, gives $y = 9$ high bacter $X$ level.

#### E. Find the frequentist estimator for $p$.

Given the new data, I proceed as in the previous points. The frequentist estimator is now $\hat{p} = y/n = 9/165 = 5.46\%$.

#### F. Calculate and plot the posterior distribution, assuming both a $Beta(1,10)$ prior for $p$ and assuming the posterior probability of the older measurement as the prior for the new one.

The prior distributions have paramters $(\alpha_1,\beta_1) = (1,10)$ and $(\alpha_2,\beta_2) = (12,115)$. Again, those are conjugate priors of the model, thus the posteriors they lead to will respectively be Beta distributions with paramters: $(\alpha_1',\beta_1') = (10,166)$ and $(\alpha_2',\beta_2') = (21,271)$.

```r
#Beta prior paramters
alpha.prior.1 = 1
beta.prior.1 = 10
alpha.prior.2 = 12
beta.prior.2 = 115

#Experiment outcomes
n = 165
y = 9

#Beta posterior parameters
alpha.post.1 = alpha.prior.1 + y
beta.post.1 = beta.prior.1 + n - y
alpha.post.2 = alpha.prior.2 + y
beta.post.2 = beta.prior.2 + n - y

#Beta posterior statistical indexes
mu.post.1 = alpha.post.1/(alpha.post.1+beta.post.1)
sigma.post.1 = sqrt((alpha.post.1*beta.post.1)/((alpha.post.1+beta.post.1+1)*(alpha.post.1+beta.post.1)**2))
mu.post.2 = alpha.post.2/(alpha.post.2+beta.post.2)
sigma.post.2 = sqrt((alpha.post.2*beta.post.2)/((alpha.post.2+beta.post.2+1)*(alpha.post.2+beta.post.2)**2))

#Frequentist estimator
p.freq = y/n

#Plot
options(repr.plot.width=12, repr.plot.height=6)

p = seq(0,0.3,len=201)
prior.1 = dbeta(p,alpha.prior.1,beta.prior.1)
prior.2 = dbeta(p,alpha.prior.2,beta.prior.2)
post.1 = dbeta(p,alpha.post.1,beta.post.1)
post.2 = dbeta(p,alpha.post.2,beta.post.2)

ylim=c(0,28)
xlim=c(0,0.3)
plot(p,prior.1,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,ylab='PDF',
     xlab='p',main='Priors and posteriors', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 2, lwd = 3)
par(new=TRUE)
plot(p,post.1,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,col='navy',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(p,prior.2,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,col='red',type='l', lty = 2, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(p,post.2,xaxs='i',yaxs='i',xlim=xlim,ylim=ylim,col='red',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
abline(v=mu.post.1,lty=3,lwd=2,col='navy')
abline(v=mu.post.2,lty=3,lwd=2,col='red')
legend('topright', c('Prior 1','Posterior 1','Prior 2','Posterior 2'), col=c("navy", "navy", "red", 'red'), lty=c(2,1,2,1))
```

As we can see the two priors lead to very different posteriors, where the mean values differ greatly.

#### G. Find the bayesian estimator for $p$, the posterior mean and variance, and a $95\%$ credible interval.

```r
q = 0.025
CI.min.1 = qbeta(q,alpha.post.1,beta.post.1)
CI.max.1 = qbeta(1-q,alpha.post.1,beta.post.1)
CI.min.2 = qbeta(q,alpha.post.2,beta.post.2)
CI.max.2 = qbeta(1-q,alpha.post.2,beta.post.2)

sprintf('The estimator of the frequentist approach is %.2f%%', 100*p.freq)
sprintf('The estimator in the Bayesian case for the first prior is %.2f%% with a standard deviation of %.2f%%.',100*mu.post.1, sigma.post.1*100)
sprintf('The estimator in the Bayesian case for the second prior is %.2f%% with a standard deviation of %.2f%%.',100*mu.post.2, sigma.post.2*100)
sprintf('The credibility interval for the first posterior is: [%.2f%%,%.2f%%]',CI.min.1*100,CI.max.1*100)
sprintf('The credibility interval for the second posterior is: [%.2f%%,%.2f%%]',CI.min.2*100,CI.max.2*100)
```

## Exercise 1 - A new journal
A publishing company has recently launched a new journal. In order to determine how effective it is in reaching its possible audience, a market survey company selects a random sample of people from a possible target audience and interviews them. Out of 150 interviewed people, 29 have read the last issue of the journal.

#### A. What kind of distribution would you assume for y, the number of people that have seen the last issue of the journal?

Assuming that the target is effectively made of suitable readers and given the fact that an interviewee could have either read or not the journal, I would say that the best model for this problem is a Binomial likelihood.
In this framework, the "Event: success" is encoded by the syntax: "has read the last issue of the journal", and the "failure" by its opposite. Hence, for the case under investigation, the result is $r=29$ successes out of $n=150$ trials.

#### B. Assuming a uniform prior, what is the posterior distribution for y ?

Let the prior be $f(p|M) = U(0,1) = Beta(1,1)$ (Uniform for $p \in [0,1]$) and the likelihood $P(r|n,p,M) = Binom(r; n, p)$, for which the Beta distribution is the conjugate prior. Writing the uniform distribution in this way as well allows to save computational resources. Hence, recalling that for a generic Beta prior $Beta(\alpha_P, \beta_P)$ and binomial likelihood the posterior will be another beta distribution with parameters $\alpha = \alpha_P + r, \beta = \beta_P + (n-r)$, where $r$ is the number of successes, while using $\alpha_P = \beta_P = 1$, $r=29$ and $n=150$ the posterior will be $Beta(\alpha=30,\beta=122)$.

#### C. Plot both posterior and likelihood ditributions functions

```r
#Define functions
mean.of.beta <- function(alpha, beta, n, r){
    (alpha + r)/(alpha + beta + n)
}

mode.of.beta <- function(alpha, beta, n, r){
    (alpha + r - 1)/(alpha + beta + n - 2)
}
```

For further use, I define functions to estimate the mean and the mode of a Beta distribution. From the theory, it holds:

$$
    mean = \frac{\alpha_p + r}{\alpha_p + \beta_p + n} \ \ \ \ \ \ \ mode = \frac{\alpha_p + r-1}{\alpha_p + \beta_p + n-2}
$$

Notice how given an experiment output and the parameters of the distribution it is straightforward to evaluate those values.

```r
options(repr.plot.width=12, repr.plot.height=6)

#Set parameters
alpha = 30
beta = 122
n = 150
r = 29
p1 = 0.5
p2 = 0.2

#Generate x axes
p = seq(0,1,len=501)
success = 1:150

#Define pdfs
prior = dbeta(p, 1, 1)
post = dbeta(p, alpha, beta)

likelihood_1 = dbinom(success, size=n, prob=p1)
likelihood_2 = dbinom(success, size=n, prob=p2)


#Plots
par(mfrow=c(1,2))

plot(p,prior,xaxs='i',yaxs='i',ylim=c(0,13),ylab='PDF',
     xlab='p',main='Prior and posterior', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
par(new=TRUE)
plot(p,post,ylim=c(0,13),xaxs='i',yaxs='i',col='navy',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
legend('topright', c('Prior','Posterior'), col=c("red", "navy"), lty=1:1, bty = 'n')

plot(success,likelihood_1,xaxs='i',yaxs='i',ylab='P(r|n,p,M)', ylim=c(0,.2),
     xlab='Number of Heads r',main=paste('Binomial likelihood for n =',n), cex.lab=1.5, cex.main=2, 
     type='h', lwd=5, col='chartreuse3')
par(new=TRUE)
plot(success,likelihood_2,xaxs='i',yaxs='i',axes=FALSE, ann=FALSE,pch=19, col='orange',  type='h', lwd=5)
abline(v=r,lty=2,col='black')
legend('topright', c(paste('p =',p1),paste('p =',p2), paste('r obs =', r)), 
       fill=c("chartreuse3", "orange", 'black'), bty = 'n')
```

```r
sprintf('The mean of the posterior is: %.4f', mean.of.beta(30,122,150,29))
sprintf('The mode of the posterior is: %.4f', mode.of.beta(30,122,150,29))
```

## Exercise 2 - The new concert hall

Three students want to construct their prior probability about the proportion of residents that support the building of a new concert hall in their small town.
- Anna thinks that her prior is a beta distribution with mean 0.2 and a standard deviation of 0.08.
- Benny moved only recently to this new town and therefore he does non have the slightest idea about
it. Therefore he decides to use a uniform prior.
- Chris believes that his prior should have a trapezoidal shape:
$$
f(X) =  \left\{ \begin{array}{4}
20x   & \ \ \ 0   \le x < 0.1 \\
2     & 0.1 \le x < 0.3 \\
5−10x & 0.3 \le x < 0.5 \\
0     & \ \ \ \ 0.5 \le x
\end{array}\right.
$$

#### A. Draw and compare the three prior distributions.

First of all let me define the trapezoidal distribution. One can notice that the function above is not normalized for $p \in [0,1]$, hence I divide the outputs for the trapezoid Area (given by the basic formula for trapezoids)

```r
Area.trap = ((0.5+0.2)*2)/2

trap_function <- function(x){
    dplyr::case_when(
    0  <= x & x <= 0.1 ~ 20*x/Area.trap,
    0.1 < x & x <= 0.3 ~ 2/Area.trap,
    0.3 < x & x <= 0.5 ~ (5-10*x)/Area.trap,
    0.3 < x            ~ 0
    )
}
dtrap <- Vectorize(trap_function)
```

Now define all functions and proceed to plot:

```r
options(repr.plot.width=10, repr.plot.height=6)

#Set x axis
p = seq(0,1,len=501)

#Anna's prior from sigma and mu
A.mu = 0.2
A.sigma = 0.08

A.alpha = (((1-A.mu)/(A.sigma**2))-(1/A.mu))*A.mu**2
A.beta = A.alpha*((1/A.mu)-1)

A.prior = dbeta(p, A.alpha, A.beta)

#Benny's prior
B.prior = dbeta(p,1,1)

#Chris' prior
C.prior = dtrap(p)

#Plot
ylim = c(0,6)

plot(p,A.prior,xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|M)',
     xlab='p',main='Prior distributions', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
par(new=TRUE)
plot(p,B.prior,ylim=ylim,xaxs='i',yaxs='i',col='navy',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(p,C.prior,ylim=ylim,xaxs='i',yaxs='i',col='chartreuse3',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
legend('topright', c('Anna','Benny','Chris'), col=c("red", "navy", "chartreuse3"), lty=1:1:1, bty='n')
```

• The next day the three students decide to interview a sample of 100 citizens of the small town, asking for their opinion. Out of the interviewed sample, 26 support the building of the new concert hall.
#### B. Evaluate and draw the three posterior distributions.

Given the fact that there are two possible outcomes of an interview ("Like" and "Dislike", linkable to the "Success" and "Failure" notions) I assume for the model a binomial likelihood. Hence, Anna's posterior is simple to find, since it holds what said in Exercise 1 for the conjugate prior. Same applies to Benny's prior, which is a $Beta(1,1)$ distribution:

```r
#Experiment parameters
n = 100
r = 26

#Anna's posterior parameters
A.post.alpha = A.alpha + r
A.post.beta = A.beta + n - r

#Benny's posterior parameters
B.post.alpha = 1 + r
B.post.beta = 1 + n - r

#Posteriors
A.posterior = dbeta(p, A.post.alpha, A.post.beta)
B.posterior = dbeta(p, B.post.alpha, B.post.beta)
```

As for Chris' posterior instead, a numerical evaluation is required. In order to do so, I calculate the normalization factor by summing over little steps $\Delta p$ such that $p_i = i \times \Delta p$ and evaluate the product of likelihood and prior in those values:

$$
Z = \int_0^1 P(r|n,p,M) P(p|M) dp = \int_0^1 Binom(r;n,p) Trap(p) dp \approx \sum_{p_i} (Binom(r;n,p_i) Trap(p_i)) \cdot \Delta p 
$$

```r
#Interval discretization
n.deltas = 1000
Delta.p = 1/n.deltas
p.i = seq(0,1,Delta.p)

#Integral approximation
C.propto.posterior <- function(p){ dbinom(26,100,p) * dtrap(p) }
Z = Delta.p * sum(C.propto.posterior(p.i))

#Renormalized posterior definition
C.posterior <- function(p){ C.propto.posterior(p) / Z }
```

```r
options(repr.plot.width=8, repr.plot.height=6)

ylim = c(0,11)

plot(p,A.posterior,xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|M)',
     xlab='p',main='Posterior distributions', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
par(new=TRUE)
plot(p,B.posterior,ylim=ylim,xaxs='i',yaxs='i',col='navy',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(p,C.posterior(p),ylim=ylim,xaxs='i',yaxs='i',col='chartreuse3',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
legend('topright', c('Anna','Benny','Chris'), col=c("red", "navy", "chartreuse3"), lty=1:1:1, bty='n')
```

As one can notice, the posteriors of Benny and Chris are quite similar, as the statistics is quite large and they both start with similar assumprions (a wide range of equally probable values). On the other hand, Anna's posterior is biased by her initial assumption, leading to a curve which is peaked in a lower region w.r.t the others. 

#### C. Give an estimate of the most probable value and the 95% credibility interval.

For Beta posteriors the mode, i.e. the most probable value, is know from the theory (look above).

To find the 95% credibility interval I look for $p_{min}, p_{max}$ such that:

$$
\int_0^{p_{min}} P(p|r,n,M) \ dp = \int_{p_{max}}^1 P(p|r,n,M) \ dp = 2.5\%
$$

And $CI = (p_{min}, p_{max})$. Thus, for the first two cases I exploit the quantile function of R for the Beta distribution:

```r
#Modes evaluation
A.mode = mode.of.beta(A.alpha,A.beta,n,r)
B.mode = mode.of.beta(1,1,n,r)

#Quantile for tails
q = 0.025

#Credibility intervals
A.p.min = qbeta(q, A.post.alpha, A.post.beta)
A.p.max = qbeta(1-q, A.post.alpha, A.post.beta)

B.p.min = qbeta(q, B.post.alpha, B.post.beta)
B.p.max = qbeta(1-q, B.post.alpha, B.post.beta)
```

For Chris instead is convenient to look for the max of the PDF (mode) and compute numerically the CDF to look for the 2.5% and 97.5% quantiles (Credibility Interval):

```r
#Interval discretization
n.deltas = 2000
Delta.p = 1/n.deltas
p.i = seq(0,1,Delta.p)

#Mode evaluation
C.mode = p.i[(C.posterior(p.i) == max(C.posterior(p.i)))]

#CDF and Credibility interval
C.CDF = cumsum(C.posterior(p.i))*Delta.p
C.CredInt = p.i[(q < C.CDF & C.CDF < 1-q)]
C.p.min = min(C.CredInt)
C.p.max = max(C.CredInt)
```

Plotting those on the previous graph highlights the correct estimations as well as what said for the peaks (which are coincident for Benny and Chris' posteriors).

```r
ylim = c(0,11)

plot(p,A.posterior,xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|M)',
     xlab='p',main='Posterior distributions and modes', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
abline(v=A.mode,lty=2,col='red')
par(new=TRUE)
plot(p,B.posterior,ylim=ylim,xaxs='i',yaxs='i',col='navy',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
abline(v=B.mode,lty=2,col='navy')
par(new=TRUE)
plot(p,C.posterior(p),ylim=ylim,xaxs='i',yaxs='i',col='chartreuse3',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
abline(v=C.mode,lty=2,col='chartreuse3')
legend('topright', c('Anna','Benny','Chris'), col=c("red", "navy", "chartreuse3"), lty=1:1:1)

sprintf('The mode for Anna posterior is: %.3f', A.mode)
sprintf('The mode for Benny posterior is: %.3f', B.mode)
sprintf('The mode for Chris posterior is: %.3f', C.mode)
```

Finally, let's have a look at the estimation of credibility intervals:

```r
#Print the CIs:
sprintf('For Anna posterior the CI is: (%.3f,%.3f)',
        A.p.min,A.p.max)
sprintf('For Benny posterior the CI is: (%.3f,%.3f)',
        B.p.min,B.p.max)
sprintf('For Chris posterior the CI is: (%.3f,%.3f)',
        C.p.min,C.p.max)

#Plots:
options(repr.plot.width=18, repr.plot.height=6)

ylim = c(0,11)

par(mfrow=c(1,3))

plot(p,A.posterior,xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|M)',
     xlab='p',main='C.I. for Anna', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
par(new=TRUE)
ci.unif = seq(A.p.min,A.p.max,len=200)
polygon(c(ci.unif ,A.p.max, A.p.min), c(dbeta(ci.unif, A.post.alpha, A.post.beta), 0, 0), col="red", bor=NA)

plot(p,B.posterior,xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|M)',
     xlab='p',main='C.I. for Benny', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 1, lwd = 3)
par(new=TRUE)
ci.unif = seq(B.p.min,B.p.max,len=200)
polygon(c(ci.unif ,B.p.max, B.p.min), c(dbeta(ci.unif, B.post.alpha, B.post.beta), 0, 0), col="navy", bor=NA)

plot(p,C.posterior(p),xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|M)',
     xlab='p',main='C.I. for Chris', cex.lab=1.5, cex.main=2,
     col='chartreuse3',type='l', lty = 1, lwd = 3)
par(new=TRUE)
ci.unif = seq(C.p.min,C.p.max,len=200)
polygon(c(ci.unif ,C.p.max, C.p.min), c(C.posterior(ci.unif), 0, 0), col="chartreuse3", bor=NA)
```

## Exercise 3 - Coin flipping

A coin is flipped n = 30 times with the following outcomes:

$$
T, T, T, T, T, H, T, T, H, H, T, T, H, H, H, T, H, T, H, T, H, H, T, H, T, H, T, H,
H, H
$$

#### A. Assuming a flat prior, and a beta prior, plot the likelihood, prior and posterior distributions for the data set.

For the arbitrary Beta prior I fix the initial parameters to $\alpha_P = \beta_P =8$.
Theoretically speaking holds what said above for the other cases, since here again the priors are Beta distributions and the model associated to coin tossing is naturally a binomial. Associating the concept of "success" to the "Head" outcome, since in this case the experiment returns $r=15$ Heads out of $n=30$ trials, the posteriors will be:

$$
P_u(p|r,n,M) = \frac{1}{Z} Binom(n=30,r=15) Beta(1,1) = Beta(16, 16) \\
P_{\beta}(p|r,n,M) = \frac{1}{Z} Binom(n=30,r=15) Beta(8,8) = Beta(23, 23)
$$

Where Z is the suitable normalization factor. The plot is then straightforward:

```r
options(repr.plot.width=15, repr.plot.height=6)

#Parameters definition
alpha.prior = 8
beta.prior = 8
r = 15
n = 30

p1 = 0.5
p2 = 0.8

#Set x axes
p = seq(0,1,len=501)
success = 1:30

#Define PDFs 
p.unif = dbeta(p,1,1)
p.beta = dbeta(p,alpha.prior,beta.prior)

likelihood_1 = dbinom(success, size=n, prob=p1)
likelihood_2 = dbinom(success, size=n, prob=p2)

a.post.unif = 1+r
b.post.unif = 1+n-r
a.post.beta = alpha.prior+r
b.post.beta = beta.prior+n-r

post.unif = dbeta(p,a.post.unif,b.post.unif)
post.beta = dbeta(p,a.post.beta,b.post.beta)

#Calculate mean and mode for the posteriors
post.mean.unif = mean.of.beta(1,1,n,r)
post.mode.unif = mode.of.beta(1,1,n,r)
post.mean.beta = mean.of.beta(alpha.prior,beta.prior,n,r)
post.mode.beta = mode.of.beta(alpha.prior,beta.prior,n,r)

#Plots
plot(success,likelihood_1,ylab='P(r|n,p,M)', ylim=c(0,.2),
     xlab='Number of Heads r',main=paste('Binomial likelihood for n =',n), cex.lab=1.5, cex.main=2, 
     type='h', lwd=4, col='chartreuse3')
par(new=TRUE)
plot(success,likelihood_2,axes=FALSE, ann=FALSE,pch=19, col='orange',  type='h', lwd=5)
abline(v=r,lty=2,col='black')
legend('topleft', c(paste('p =',p1),paste('p =',p2), paste('r obs =', r)), 
       col=c("chartreuse3", "orange", 'black'), lty=c(1,1,2),bty = 'n')

par(mfrow=c(1,2))

plot(p,p.beta,xaxs='i',yaxs='i',ylim=c(0,3.4),ylab='P(p|M)',
     xlab='p',main='Prior distributions', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
par(new=TRUE)
plot(p,p.unif,ylim=c(0,3.4),xaxs='i',yaxs='i',col='navy',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
legend('topleft', c(paste('Beta(',alpha.prior,',',beta.prior,')'),'Flat'), col=c("red", "navy"), lty=1:1,bty = 'n')

plot(p,post.unif,xaxs='i',yaxs='i',ylim=c(0,6),ylab='P(p|n,r,M)',
     xlab='p',main='Posterior distributions', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 1, lwd = 3)
abline(v=post.mean.unif,lty=2,col='red')
par(new=TRUE)
plot(p,post.beta,xaxs='i',yaxs='i',ylim=c(0,6),col='red',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
abline(v=post.mean.beta,lty=3,col='navy')
legend('topleft', c('Beta prior','Flat prior'), col=c("red", "navy"), lty=1:1, bty = 'n')
```

From the plots above we can see that really different priors lead to verry similarly centered posteriors, even if the shape is different.

#### B.  Evaluate the most probable value for the coin probability p and, integrating the posterior probability distribution, give an estimate for a 95% credibility interval.

The modes are calculated by the means of the previously defined functions and are plotted in the graph above. In particular, we can see that the two coincide:

```r
sprintf('The mode of the Beta posterior with flat prior is: %.3f', post.mode.unif)
sprintf('The mode of the Beta posterior with Beta prior is: %.3f', post.mode.beta)
```

As for the credibility intervals instead:

```r
#Set the quantile:
q = 0.025

#Find the thresholds:
p.min.unif = qbeta(q,a.post.unif,b.post.unif)
p.max.unif = qbeta(1-q,a.post.unif,b.post.unif)
p.min.beta = qbeta(q,a.post.beta,b.post.beta)
p.max.beta = qbeta(1-q,a.post.beta,b.post.beta)
sprintf('For the Beta posterior with flat prior the CI is (%.3f,%.3f)',
        p.min.unif,p.max.unif)
sprintf('For the Beta posterior with Beta prior the CI is: (%.3f,%.3f)',
        p.min.beta,p.max.beta)


#Plot the results:
par(mfrow=c(1,2))

plot(p,post.unif,xaxs='i',yaxs='i',ylim=c(0,6),ylab='P(p|M)',
     xlab='p',main='C.I. for Beta with flat prior', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 1, lwd = 3)
par(new=TRUE)
ci.unif = seq(p.min.unif,p.max.unif,len=200)
polygon(c(ci.unif ,p.max.unif, p.min.unif), c(dbeta(ci.unif, a.post.unif, b.post.unif), 0, 0), col="navy", bor=NA)

plot(p,post.beta,xaxs='i',yaxs='i',ylim=c(0,6),ylab='P(p|n,r,M)',
     xlab='p',main='C.I. for Beta with Beta prior', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
par(new=TRUE)
ci.beta = seq(p.min.beta,p.max.beta,len=200)
polygon(c(ci.beta ,p.max.beta, p.min.beta), c(dbeta(ci.beta, a.post.beta, b.post.beta), 0, 0), col="red", bor=NA)
```

As expected, since the second posterior is more peaked the credibiity interval is narrower. 

#### C.  Repeat the same analysis assuming a sequential analysis of the data. Show how the most probable value and the credibility interval change as a function of the number of coin tosses (i.e. from 1 to 30).

Starting from the initial priors, we now toss the coin once per time and update the prior w.r.t the knowledge we acquire by the event. As a result, in $n=30$ tosses we will update the posterior $n$ times, since each time the posterior at iteration $i$ will represent our knowledge of the problem and will thus serve as prior for the iteration $i+1$. This will eventually mean make use also of the history of the draws, not only the final output.

```r
#Define data
T = 0
H = 1
data = c(T, T, T, T, T, H, T, T, H, H, T, T, H, H, H, T, H, T, H, T, H, H, T, H, T, H, T, H, H, H)

#Set x axes
p = seq(0,1,len=501)
n.seq = 0:30

#Set inital parameters
n = 0
r = 0

#Set values for unif prior
alpha.unif = 1 
beta.unif = 1
p.min.unif = qbeta(q,alpha.unif,beta.unif)
p.max.unif = qbeta(1-q,alpha.unif,beta.unif)
mode.unif = mode.of.beta(alpha.unif,beta.unif,n,r)

#Set values for beta prior
alpha.beta = 8
beta.beta = 8
p.min.beta = qbeta(q,alpha.beta,beta.beta)
p.max.beta = qbeta(1-q,alpha.beta,beta.beta)
mode.beta = mode.of.beta(alpha.beta,beta.beta,n,r)

#Set quantile for CI
q = 0.025

#Loop over events
for (event in data){
    #Update
    n =  1
    r =  event
    
    #Find params of new beta posterior
    alpha.unif = alpha.unif + r
    beta.unif = beta.unif + n - r
    alpha.beta = alpha.beta + r
    beta.beta = beta.beta + n - r
    
    #Calculate new CI and mode
    p.min.unif = c(p.min.unif,qbeta(q,alpha.unif,beta.unif))
    p.max.unif = c(p.max.unif,qbeta(1-q,alpha.unif,beta.unif))
    p.min.beta = c(p.min.beta,qbeta(q,alpha.beta,beta.beta))
    p.max.beta = c(p.max.beta,qbeta(1-q,alpha.beta,beta.beta))
    
    mode.unif = c(mode.unif, mode.of.beta(alpha.unif,beta.unif,n,r))
    mode.beta = c(mode.beta, mode.of.beta(alpha.beta,beta.beta,n,r))
}


#Plot
plot(n.seq,mode.unif,ylim=c(0,1),ylab='p',
     xlab='n',main='C.I. and mode for iterative posteriors', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 1, lwd = 3)
par(new=TRUE)
plot(n.seq,p.min.unif,ylim=c(0,1),
     col='navy',type='l', lty = 2, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(n.seq,p.max.unif,ylim=c(0,1),
     col='navy',type='l', lty = 2, lwd = 3, axes=FALSE, ann=FALSE)
polygon(c(n.seq,rev(n.seq)), c(p.max.unif,rev(p.min.unif)), col=adjustcolor("navy",alpha.f=0.2), bor=NA)
par(new=TRUE)
plot(n.seq,mode.beta,ylim=c(0,1),
     col='red',type='l', lty = 1, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(n.seq,p.min.beta,ylim=c(0,1),
     col='red',type='l', lty = 2, lwd = 3, axes=FALSE, ann=FALSE)
par(new=TRUE)
plot(n.seq,p.max.beta,ylim=c(0,1),
     col='red',type='l', lty = 2, lwd = 3, axes=FALSE, ann=FALSE)
polygon(c(n.seq,rev(n.seq)), c(p.max.beta,rev(p.min.beta)), col=adjustcolor("red",alpha.f=0.1), bor=NA)
legend('topright', c('Flat prior','Beta prior'), fill=c("navy", "red"), bty = 'n', cex=1.5)
grid()
```

This graph shows how the projection of the confidence intervals and modes evolve at each iteration. As for the flat prior, we observe how evolution turns the wide CI into a narrower range, while the mode changes in a sudden way. The Beta prior instead evolves keeping a similar shape, and the intial bias towards a smaller probability of Head moves back to 50% at the final iteration. In the end, both modes converge to similar values and CIs. 

#### D.  Do you get a different result, by analyzing the data sequentially with respect to a one-step analysis (i.e. considering all the data as a whole)?

```r
#Final distributions for sequential analysis (final alpha and beta updates)
seq.posterior.unif = dbeta(p,alpha.unif,beta.unif)
seq.posterior.beta = dbeta(p,alpha.beta,beta.beta)

#Plots
ylim=c(0,6)
par(mfrow=c(1,2))

plot(p,post.unif,xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|n,r,M)',
     xlab='p',main='Posteriors for flat prior', cex.lab=1.5, cex.main=2,
     col='navy',type='l', lty = 1, lwd = 3)
par(new=TRUE)
plot(p,seq.posterior.unif,xaxs='i',yaxs='i',ylim=ylim,col='navy',type='h', axes=FALSE, ann=FALSE)
legend('top', c('Border: One-step analysis','Color: Sequential analysis'),  bty = 'n')

plot(p,post.beta,xaxs='i',yaxs='i',ylim=ylim,ylab='P(p|n,r,M)',
     xlab='p',main='Posteriors for Beta prior', cex.lab=1.5, cex.main=2,
     col='red',type='l', lty = 1, lwd = 3)
par(new=TRUE)
plot(p,seq.posterior.beta,xaxs='i',yaxs='i',ylim=ylim,col='red',type='h', axes=FALSE, ann=FALSE)
legend('top', c('Border: One-step analysis','Color: Sequential analysis'),  bty = 'n')
```

As the plot above shows, the last iteration posterior matches perfectly the one of the one-step analysis.

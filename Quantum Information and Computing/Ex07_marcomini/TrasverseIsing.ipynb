```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import eigenvals 
N = 5
Ls = [0,1,2,3]

eigenvals0 = np.genfromtxt('/Users/ale/QuantumComputingStuff/Ex07_marcomini/data/Eigenvals_N'+str(N)+'_L0.csv', delimiter=',')
eigenvals1 = np.genfromtxt('/Users/ale/QuantumComputingStuff/Ex07_marcomini/data/Eigenvals_N'+str(N)+'_L1.csv', delimiter=',')
eigenvals2 = np.genfromtxt('/Users/ale/QuantumComputingStuff/Ex07_marcomini/data/Eigenvals_N'+str(N)+'_L2.csv', delimiter=',')
eigenvals3 = np.genfromtxt('/Users/ale/QuantumComputingStuff/Ex07_marcomini/data/Eigenvals_N'+str(N)+'_L3.csv', delimiter=',')

EIGENVALS = [eigenvals0,eigenvals1,eigenvals2,eigenvals3]
k = 2**(N-1)
print(k)

for i in Ls:
    plt.scatter(np.arange(k),EIGENVALS[i][0:k]/N, label=str(i), s=100/N)
    plt.plot(np.arange(k),EIGENVALS[i][0:k]/N, ls=':')

plt.legend(title='$\lambda$')
plt.title('First %i eigenvals for TFI (N=%i)' % (k,N))
plt.xlabel('order')
plt.ylabel('E/N')
plt.grid()
plt.show()
```
